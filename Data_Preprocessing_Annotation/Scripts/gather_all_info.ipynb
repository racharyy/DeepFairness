```python
import numpy as np, pandas as pd, pickle
import torch
```

```python
def load_pickle(pickle_file):
    try:
        with open(pickle_file, 'rb') as f:
            pickle_data = pickle.load(f)
    except UnicodeDecodeError as e:
        with open(pickle_file, 'rb') as f:
            pickle_data = pickle.load(f, encoding='latin1')
    except Exception as e:
        print('Unable to load data ', pickle_file, ':', e)
        raise
    return pickle_data
```

```python
doc_rep = load_pickle('../Ted_data/doc_rep.pkl')
```

```python
d1 = load_pickle('../TED_data/views_ratings.pkl')
```

```python
views_ratings = pd.DataFrame(d1, index=None)
```

```python
new_cols_order = ['Video_ID',
 'totalviews',
 'total_count',
 'beautiful',
 'confusing',
 'courageous',
 'fascinating',
 'funny',
 'informative',
 'ingenious',
 'inspiring',
 'jaw-dropping',
 'longwinded',
 'obnoxious',
 'ok',
 'persuasive',
 'unconvincing']
```

```python
view_rating_df = views_ratings[new_cols_order]
```

```python
doc_rep_df = pd.DataFrame({'Video_ID': view_rating_df['Video_ID'], 'doc_rep':doc_rep})
```

```python
protected_df = pd.read_csv('../Ted_data/protected_features.csv')
```

```python
rating_rep = pd.merge(view_rating_df, doc_rep_df, on='Video_ID', how = 'inner')
```
